```python
from statsmodels.tsa.stattools import adfuller
```

```python
def test_stationarity(example):
    dftest = adfuller(example,autolag = 'AIC')
    return dftest[1]
```

上面的函数是一个简易的ADF检验函数，以AIC赤池信息准则为判断统计量。将数据导入该函数后，导出的是prob,即不通过ADF的概率，若直接使用adfuller函数，则返回p-value和90%，95%，99%三个临界值，并返回prob.通常情况下，以prob小于0.05为标志，在一些情况下，prob小于0.1也可接受。

```python
arma_mod10 = sm.tsa.ARMA(example3,(1,0)).fit()
resid = arma_mod10.resid
```

```python
resid
```

残差序列resid是最后一个需要检查的东西，它代表了我们扔掉的信息。计量模型假设残差是白噪声，也就是说数据提供的所有信息都被模型利用。所以在最后检查resid，确定我们是不是需要返工，画个图，检查一下自相关性，顺便检查一下正态性也可以。

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
plt.show()
```

两张图上，滞后项的小柱子没有超过蓝色区域的，说明resid基本可以看成白噪声了。

```python
print(sm.stats.durbin_watson(resid.values))
```

在检验自相关性质的时候，杜宾-沃特森方法不是太好的一种方式，因为它不精确，不过这里应该够用了。DW检验量取值为[0,4]，当检验量趋近与0或者4时，代表这个序列有强自相关性（0处为正相关，4处为负相关），当检验量趋近2时，则视作无自相关性。说DW不精确是因为当检验量不位于这三个值周边时，其结果很难判断。不过Durbin和Watson给出了可判读范围的计算方法，当统计量落在可判读范围内时，可以进行判断。DW方法用来做粗略的判断，可以通过其他方法来细致地判断其自相关性。这里的DW=1.914,很接近2，所以残差可以视作没有自相关性。
